```python
import pint.orbital.kepler as kepler
```

```python
p = kepler.KeplerTwoBodyParameters(
    a=7, pb=1, eps1=0, eps2=0.5, 
    i=0, lan=0, q=1, 
    x_cm=0, y_cm=0, z_cm=0,
    vx_cm=0, vy_cm=0, vz_cm=0,
    tasc=0)
ts, _ = kepler.kepler_two_body(p, 0)
print ts[6], ts[13]
```

```python
ts = np.linspace(0,p.pb,1000)
dilation = [kepler.time_dilation(kepler.kepler_two_body(p, t)[0]) for t in ts]
plt.plot(ts, dilation)
```

```python
from scipy.integrate import odeint
```

```python
r = odeint(lambda x,t: kepler.time_dilation(kepler.kepler_two_body(p, t)[0]), 
           0, ts, atol=1e-14)[:,0]
plt.plot(ts,r)
plt.title("Accumulated delay")
plt.figure()
slope = r[-1]-r[0]
plt.plot(ts,r-slope*ts)
plt.title("Trend-removed Einstein delay")
```

Our goal is to produce formulae that can replicate both the slope and the details of the above curves. Math, therefore, follows.

If $\gamma(t)\geq 1$ is the slowing factor at time $t$, the quantity we actually want is
$$
d(T) = \int_0^T (\gamma^{-1}(t)-1) dt.
$$
Unfortunately the formula for $\gamma(t)$ is rather messy. Choose instead $\theta$ to be the true anomaly corresponding to $t$, and then we can write $r(\theta)=a(1-e^2)/(1+e \cos\theta)$, where $a$ is the semimajor axis of the total ellipse - that is, half the distance between bodies at their furthest. The velocity is a little messier; in the reduced problem, $v^2 = 2\mu($... blerg.

```python
mean_anomaly = 2*np.pi*ts
eccentric_anomaly = np.array([kepler.eccentric_from_mean(np.hypot(p.eps1,p.eps2), m)[0] for m in mean_anomaly])
t2_gamma = 2e-8
t2_einstein_delay = t2_gamma*np.sin(eccentric_anomaly)
plt.plot(ts, -t2_einstein_delay)
plt.plot(ts,r-slope*ts)
```
